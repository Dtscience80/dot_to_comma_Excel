```python
import pandas as pd

# Baca file Excel dengan separator desimal titik (.)
input_excel_file = '/content/DA[10]_CD.xlsx'
df = pd.read_excel(input_excel_file)

# Mengganti header kolom dengan header yang tidak memiliki karakter '(1)'
new_columns = [col.replace('(1)', '').strip() for col in df.columns]
df.columns = new_columns

# Fungsi untuk mengubah nilai sel ke tipe float jika berupa angka
def convert_to_float(value):
    try:
        return float(value.replace(',', '.'))
    except (ValueError, AttributeError):
        return value

# Cek apakah kata 'continuous' ada di baris ke-2 dan kata 'meta' ada di baris ke-3
if df.iloc[1, 0] == 'continuous' and df.iloc[2, 0] == 'meta':
    # Hapus baris ke-2 dan ke-3
    df = df.drop([1, 2])

# Reset indeks setelah menghapus baris
df = df.reset_index(drop=True)

# Ubah separator desimal ke koma (,) dan konversi ke tipe float jika berupa angka
df = df.applymap(lambda x: convert_to_float(str(x)) if isinstance(x, (int, float, str)) else x)

# Membagi string menjadi dua bagian, sebelum dan setelah ".xlsx"
parts = input_excel_file.rsplit(".xlsx", 1)

# Menambahkan kata 'fix' ke bagian pertama dan kemudian menggabungkannya kembali dengan bagian kedua
output_string = f"{parts[0]}_fix.xlsx"

# Simpan dataframe ke file Excel baru dengan separator desimal koma (,) dan tipe float
output_excel_file = output_string
df.to_excel(output_excel_file, index=False, float_format='%.2f')

```
